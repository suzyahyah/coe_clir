```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np
```

```python
lang="SOMA"
models=["tm", "human", "mt1", "mt2"]

all={}



for model in models:
    with open(f'results/each_map_{lang}.{model}' ,'r') as f:
        res = f.readlines()
    
    res = [r.strip().split() for r in res]    
    queries = [r[0] for r in res]
    scores = [np.float(r[1]) for r in res]
    
    all[model] = scores
    #all['qid'] = queries
    #all['model'] = ['tm']*len(queries)

all_res = pd.DataFrame(all)
```

```python
sns.pairplot(all_res, markers=".")
```
